#1) Importing related libraries and uploading the image



```python
import matplotlib.pyplot as plt
import cv2


!wget https://i.stack.imgur.com/iIzNa.jpg
```

#2) Displaying the image

and also making sure that the image has the color channels in the right order

```python
image = cv2.imread("iIzNa.jpg")
image = image[:, :, ::-1] # ---> BGR to RGB convertion


plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```

As it seems, this image suffers from low intensities (too dark).

#3) Converting image to grayscale

```python
gray_image = cv2.imread("iIzNa.jpg", cv2.IMREAD_GRAYSCALE)
```

#4) Applying the Intensity Transformation by Power Law

Since the image is too dark, γ (gamma) value should be less than 1.

```python
# Applying Power Law
#s = c * (r**gamma)


c = 1
gamma = 0.4


transformated_image = c * (gray_image**gamma)
```

After several trials, it was decided that the appropriate value of the γ was 0.4.

#5) Plotting both of the images

```python
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(gray_image, cmap = "gray")
plt.title("Before Power Law Transformation")
plt.subplot(122)
plt.imshow(transformated_image, cmap = "gray")
plt.title("After Power Law Transformation")
plt.show()
```
